```python
#scatterplot to see correlation of each variable with calories
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,10))
#weights vs calories
axes[0,0].scatter(train_df['Weight'], train_df['Calories'], c='mediumslateblue')
axes[0,0].set_title('Weight vs Calories')
axes[0,1].scatter(train_df['Height'], train_df['Calories'], c='slateblue')
axes[0,1].set_title('Height vs Calories')
axes[0,2].scatter(train_df['Heart_Rate'], train_df['Calories'], c='darkslateblue')
axes[0,2].set_title('Heart Rate vs Calories')
axes[1,0].scatter(train_df['Age'], train_df['Calories'], c='hotpink')
axes[1,0].set_title('Age vs Calories')
axes[1,1].scatter(train_df['Duration'], train_df['Calories'], c='deeppink')
axes[1,1].set_title('Duration vs Calories')
axes[1,2].scatter(train_df['Body_Temp'], train_df['Calories'], c='mediumvioletred')
axes[1,2].set_title('Body Temp vs Calories')
plt.tight_layout()
plt.show()
```
